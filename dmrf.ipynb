```python
import requests
offset_ = 1
limit_ = 100
url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit={limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
res = requests.get(url)
```

```python
import pandas as pd
data = pd.json_normalize(res.json().get('data').get('list'))  # df для начального списка с общей информацией
data = data.rename(columns={'objId':'id'})

print(data.shape)
data.head()
```

```python
new_data = pd.DataFrame() # df для детальной информации по каждому дому

ids = data['id']
for i in range(len(ids)):
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{ids[i]}'
    res = requests.get(url)
    new_data = new_data.append(pd.json_normalize(res.json().get('data')))

print(new_data.shape)
new_data.head()
```

```python
final_data = new_data.merge(data, on='id', how='left')
print(final_data.shape)
final_data.head()
```

## photoRenderDTO

```python
def get_photoRenderDTO(column):
    """тк в некоторых полях таблицы хранится list(dict) надо привести эти поля к dataFrame"""
    ret = pd.DataFrame()
    for elem in column:
        for i in range(len(elem)):
            ret = ret.append(pd.DataFrame.from_records(data = elem[i], index=[1]))
    return ret
```

```python
photoRenderDTO = get_photoRenderDTO(final_data['photoRenderDTO'])
photoRenderDTO = photoRenderDTO.rename(columns={'objId':'id'})
photoRenderDTO
```

## objectTransportInfo

```python
def get_dict(dictionary):
    """возвращает словарь листы которого заменены на строки с сепаратором ','"""
    ret = {}
    for k in dictionary.keys():
        if type(dictionary[k]) is list:
            ret[k] = ','.join([str(item) for item in dictionary[k]])
        else:
            ret[k] = str(dictionary[k])
    return ret
```

```python
def get_objectTransportInfo(columns, column_name, id):
    """тк в некоторых полях таблицы хранится list(dict) надо привести эти поля к dataFrame.
    Кроме того необходимо добавить id к полученному DataFrame"""
    ret = pd.DataFrame()
    q = columns[column_name].copy()
    q[id] = columns[id]
    for i in range(len(columns)):
        for j in range(len(columns.iloc[i]['objectTransportInfo'])):
            z = get_dict(columns.iloc[i]['objectTransportInfo'][j])
            z['id'] = columns.iloc[i][id]
            ret = ret.append(pd.DataFrame.from_records(data = z, index=[1]))
    return ret
```

```python
objectTransportInfo = get_objectTransportInfo(final_data[['objectTransportInfo','id']], 'objectTransportInfo', 'id')
objectTransportInfo
```

## MetroColors

```python
def get_metroColors(column):
    """тк в некоторых полях таблицы хранится list(dict) надо привести эти поля к dataFrame"""
    ret = pd.DataFrame()
    for elem in column:
        for i in range(len(elem)):
            ret = ret.append(pd.DataFrame.from_records(data = elem[i], index=[1]))
    return ret
```

```python
metroColors = get_metroColors(final_data['photoRenderDTO'])
metroColors = metroColors.rename(columns={'objId':'id'})
metroColors
```
